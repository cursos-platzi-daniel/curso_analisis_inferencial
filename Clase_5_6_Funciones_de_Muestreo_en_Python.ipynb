# Funciones de muestreo:

### 1) Muestreo aleatorio simple: 
Selecciona elementos de una población **al azar**, donde todos tienen la **misma probabilidad** de ser elegidos.    
Ejemplo: Elegir 10 nombres de un sombrero.

```python
# Seleccionamos una muestra de 8 registros de la tabla de manera aleatoria
aleat_8 = econdata.sample(n=8)
aleat_8
```

```python
# Repetimos el ejercicio y observamos que la selección aleatoria cambió
aleat_8_2 = econdata.sample(n=8)
aleat_8_2
```

```python
# Otra forma de sacar una muestra es indicando un porcentaje del total de registros
# Ejemplo selección aleatoria del 25% de la tabla
prop_25 = econdata.sample(frac =.25)
prop_25.head()
```

### 2) Muestreo sistemático:
Selecciona elementos en intervalos regulares, después de un punto de inicio aleatorio.  
Ejemplo: Escoger cada 3 registros de una tabla

```python
def systematic_sampling(econdata, step):
    # genera una secuencia de índices desde 0 hasta el tamaño del dataset (len(econdata)), avanzando en pasos de tamaño step
    indexes = np.arange(0, len(econdata), step=step)
    # iloc selecciona las filas en las posiciones indicadas por indexes
    systematic_sample = econdata.iloc[indexes]
    # se devuelve como un nuevo DataFrame o Serie 
    return systematic_sample

# asignamos nuevo dataframe resulado de ejecutar la función systematic_sampling(dataframe, step)
systematic_sample = systematic_sampling(econdata, 3)

systematic_sample
```

### 3) Muestreo estratificado:
Divide la población en **grupos homogéneos (estratos)** y selecciona muestras al azar dentro de cada grupo.  
Ejemplo 1: Escoger estudiantes de cada grado escolar.  
Ejemplo 2: Escoger hoteles de cada municipio o ciudad.

value_counts(): Cuenta cuántas veces aparece cada valor único en la columna estratificado.  
len(econdata): Divide estas cuentas por el tamaño total del DataFrame para obtener proporciones.  
.sort_values(ascending=False) : Ordena de forma descendente

A continuación, se define una función "data_estratificada" que realiza un muestreo estratificado sobre un DataFrame (econdata) basándose en valores específicos de una columna y las proporciones definidas por el usuario. En este caso, la columna escogida es la que acabamos de crear "Estratificado".  


```python
def data_estratificada(econdata, nombres_columnas_estrat, valores_estrat, prop_estrat, random_state=None):

    df_estrat = pd.DataFrame(columns = econdata.columns) # Creamos un data frame vacío con los nombres de las columnas de econdata

    pos = -1 #Ayuda a identificar si estás en la última iteración del bucle
    for i in range(len(valores_estrat)): # iteración sobre los valores únicos estratificados
        pos += 1 # incremento pos = pos + 1
        if pos == len(valores_estrat) - 1: # Si es la última posición, comprueba si el índice actual (pos) es igual al índice de la última posición de la tabla
            ratio_len = len(econdata) - len(df_estrat) 
            # Si estoy en la ultima posición, ratio es el numero de filas que me falta, es decir la resta del total - el acumulado
            # Esto compensa errores de redondeo al calcular el tamaño de las muestras en iteraciones anteriores
        else: # No es la última iteración
            ratio_len = int(len(econdata) * prop_estrat[i]) # Calcula cuántas filas se deben tomar para este estrato, multiplicando el tamaño total del DataFrame por la proporción definida para ese estrato

        df_filtrado = econdata[econdata[nombres_columnas_estrat] ==valores_estrat[i]] # filtra los datos de origen según los valores seleccionados en la estratificación de datos
        
        df_temp = df_filtrado.sample(replace=True, n=ratio_len, random_state=random_state) # haz un sample de los datos filtrados usando la ratio que hemos calculado 

        df_estrat = pd.concat([df_estrat, df_temp]) # Toma las filas actuales en df_estrat (el acumulado) y agrega ó anexa las filas de df_temp y el resultado se asigna a un nuevo DataFrame con el nuevo acumulado.

    return df_estrat # Devuelve un nuevo DataFrame con la revisión total
```

```python
valores_estrat = ['Cuautémoc,Hotel', 'Cuautémoc,Museo', 'Venustiano Carranza,Hotel', 'Cuauhtémoc,Mercado','Venustiano Carranza,Mercado']
prop_estrat = [0.5, 0.2, 0.1, 0.1,0.1]
df_estrat = data_estratificada(econdata, 'estratificado', valores_estrat, prop_estrat, random_state=42)
df_estrat
```
